6. A factory manufactures two products A and B. Product A requires 2 hours of machine time and 3 kg of
raw material per unit. Product B requires 4 hours of machine time and 2 kg of raw material per unit.
Available machine time is 40 hours and raw material is 36 kg. Profit per unit is ■50 for A and ■70 for B.
Formulate the LPP and solve it graphically.


```python
import numpy as np
import matplotlib.pyplot as plt

A = np.linspace(0, 20, 400)

B1 = (40 - 2*A) / 4      # Machine constraint
B2 = (36 - 3*A) / 2      # Raw material constraint

plt.plot(A, B1, label="2A + 4B = 40")
plt.plot(A, B2, label="3A + 2B = 36")
plt.axhline(0)
plt.axvline(0)

plt.fill_between(A, 0, np.minimum(B1, B2),
                 where=(B1 >= 0) & (B2 >= 0), alpha=0.3)

# Correct corner points
points = [(0,0), (0,9), (8,6), (20,0)]
profits = [50*x + 70*y for x,y in points]

optimal = points[np.argmax(profits)]

for x,y in points:
    plt.scatter(x,y)

plt.scatter(*optimal, s=100)
plt.text(optimal[0]+0.2, optimal[1]+0.2, "Max Profit = ₹820")

plt.xlabel("Product A")
plt.ylabel("Product B")
plt.legend()
plt.show()

print("Optimal solution:")
print("x =", optimal[0], ", y =", optimal[1])
print("Maximum profit = ₹820")
```
